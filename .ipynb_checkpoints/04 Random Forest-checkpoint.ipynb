<h1>Breve Introdução</h1>

<p style="text-align:justify">
O Airbnb começou em 2008, quando dois designers que tinham um espaço sobrando hospedaram três viajantes que procuravam um lugar para ficar. Agora, milhões de anfitriões e viajantes optam por criar uma conta gratuita no Airbnb para que possam anunciar seu espaço e reservar acomodações únicas em qualquer lugar do mundo. <p><b>Medindo a Relação entre Variáveis</b></p>

<p style="text-align:justify">
Muitas vezes quando trabalhamos com análise de dados é de grande utilidade medirmos a relação entre duas variáveis, este conceito serve de base para criação de alguns tipos de modelos preditivos, explicando de forma simplificada: se sabemos a relação entre a variável X e a variável Y conseguimos criar um modelo matemático que possuindo apenas novas variáveis de entrada X irá prever a variável de saída Y.
</p>

<p style="text-align:justify">
A medida da força da correlação é expressa como um coeficiente, chamado de coeficiente de correlação de Pearson e ele assume valores entre -1 e 1.
</p>

<ul>
    <li>0.9 a 1.0 positivo ou negativo indica uma correlação muito forte;</li>
    <li>0.7 a 0.9 positivo ou negativo indica uma correlação forte;</li>
    <li>0.5 a 0.7 positivo ou negativo indica uma correlação moderada;</li>
    <li>0.3 a 0.5 positivo ou negativo indica uma correlação fraca;</li>
    <li>0 a 0.3 positivo ou negativo indica uma correlação desprezível;</li>
</ul>

<p>Vamos agora explorar a relação entre as variáveis do nosso conjunto de dados:</p>

```python
#gerando uma tabela com a correlação entre todas as variáveis do nosso dataset.
airbnb.corr()
```

<p>Outra forma visual de verificarmos a correlação entre as variáveis:</p>

```python
def plot_corr(df, size=8):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize = (3*size, 3*size))
    ax.matshow(corr)  
    plt.xticks(range(len(corr.columns)), corr.columns) 
    plt.yticks(range(len(corr.columns)), corr.columns)
```

```python
plot_corr(airbnb)
```

<p style="text-align:center; font-size:20px; font-weight:700; font-family: monospace; background-color:rgb(50,50,50); color:rgb(0,255,0); padding:1vh 1vw; border:hidden; border-radius:3px">Separação em treinamento e teste</p>

```python
strat_train_set, strat_test_set = train_test_split(airbnb, test_size=0.2, random_state=42)
```

```python
strat_test_set.dropna().shape
```

```python
strat_train_set.dropna().shape
```

```python
airbnb = strat_train_set.dropna().drop("price", axis=1)
airbnb_labels = strat_train_set.dropna()["price"].copy()
```

```python
airbnb.head(2)
```

```python
airbnb.shape, airbnb_labels.shape
```

```python
cat_attribs = ["neighbourhood_group","neighbourhood","room_type"]
num_attribs = list(airbnb.drop(columns=cat_attribs).columns)


num_pipeline = Pipeline([
    ( 'selector', DataFrameSelector(num_attribs) ),
    ( 'imputer', SimpleImputer(strategy="median")),
    ( 'std_scaler', StandardScaler() )
])
cat_pipeline = Pipeline([
    ( 'selector', DataFrameSelector(cat_attribs) ),
    ( 'label_encoder', OneHotEncoder(handle_unknown="ignore"))
])
full_pipeline = ColumnTransformer([
    ("num_pipeline", num_pipeline, num_attribs),
    ("cat_pipeline", cat_pipeline, cat_attribs),
])
```

```python
airbnb_prepared = full_pipeline.fit_transform(X=airbnb)
```

```python
airbnb_prepared.shape
```

```python
some_data = airbnb.iloc[:5]
```

```python
some_labels = airbnb_labels.iloc[:5]
```

```python
some_data_prepared = full_pipeline.transform(some_data)
```

<hr><p style="text-align:center; font-size:20px; font-weight:700; font-family: monospace;">Regressão Linear</p><hr>

```python
time_init = time()

lin_reg = LinearRegression(n_jobs=-1)
lin_reg.fit(airbnb_prepared, airbnb_labels)

time_exit = time()
print("Tempo gasto: {:.2f} segundos".format(time_exit - time_init))
```

```python
print("Predictions:\t", lin_reg.predict(some_data_prepared))
print("Labels:\t\t", list(some_labels))
```

```python
linear_prediction = lin_reg.predict(airbnb_prepared)
```

```python
lin_mse = mean_squared_error(airbnb_labels, linear_prediction)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
time_init = time()
linear_r2 = r2_score(airbnb_labels, linear_prediction)
print("\nTempo gasto: {:.2f} segundos".format(time() - time_init))
```

```python
display_scores(linear_r2)
```

```python
plt.plot(airbnb_labels, airbnb_labels, 'r-', markersize=2)
plt.plot(airbnb_labels, linear_prediction, '.', markersize=2, alpha=0.3)
plt.grid(True)
plt.show()
```

<hr><p style="text-align:center; font-size:20px; font-weight:700; font-family: monospace">Random Forest</p><hr>

```python
time_init = time()

forest_reg = RandomForestRegressor(criterion='mse', n_jobs=-1, verbose=1)
forest_reg.fit(airbnb_prepared, airbnb_labels)

print("Tempo gasto: {:.2f} segundos".format(time() - time_init))
```

```python
print("Predictions:\t", list(forest_reg.predict(some_data_prepared)) )
print("Labels:\t\t", list(some_labels))
```

```python
forest_predictions = forest_reg.predict(airbnb_prepared)
```

```python
forest_mse = mean_squared_error(airbnb_labels, forest_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse  # root mean squared error | quanto menor, melhor.
```

```python
time_init = time()
forest_r2 = r2_score(airbnb_labels, forest_predictions)
print("\nTempo gasto: {:.2f} segundos".format(time() - time_init))
```

```python
display_scores(forest_r2)
```

```python
plt.plot(airbnb_labels, airbnb_labels, 'r-', markersize=2, label="ideal")
plt.plot(airbnb_labels, forest_predictions, '+', markersize=2, alpha=0.7, label="modelo")
plt.legend()
plt.grid(True)
plt.show()
```

```python
airbnb.plot(kind="scatter", x='longitude', y='latitude', marker="o",
            s=(forest_predictions)/20, grid=True, colorbar=True, c=np.log(forest_predictions), 
            cmap=plt.get_cmap("jet"), figsize=(10,10), label="price", alpha=0.4)

plt.imshow(new_york, alpha=0.4, cmap=plt.get_cmap("jet"), extent=[-74.27, -73.67, 40.495, 40.92])
plt.xlabel("Longitude")
plt.legend()
plt.title("")
plt.savefig("RandomForest.png")
plt.show()
```

```python
forest_reg.get_params()
```

```python
# Número de árvores na random forest
n_estimators = [int(x) for x in np.linspace(start = 5, stop = 100, num = 5)]
# Numero de features a serem consideradas em cada split
max_features = ['auto', 'sqrt']
# máximo nível de profundidade de cada árvore
max_depth = [int(x) for x in np.linspace(5, 45, 9)]
max_depth.append(None)
# Número mínimo de amostras para separar cada nó
min_samples_split = [2, 3, 4, 5]
# Número mínimo de amostras para cada nó 
min_samples_leaf = [1, 2, 3, 4, 5]
# Método de seleção de amostras para trainamento de cada árvore.
bootstrap = [True, False]
# Criação dos parâmetros a serem testados
random_params = {'bootstrap': bootstrap,
                 'max_depth': max_depth,
                 'max_features': max_features,
                 'min_samples_leaf': min_samples_leaf,
                 'min_samples_split': min_samples_split,
                 'n_estimators': n_estimators,
                 'verbose': [0,1,2],
                 'warm_start': [True, False]}
```

`estimator:`modelo utilizado a ser utilziado;<br>
`param_distribution:` dicionário com parâmetros a serem testados;<br>
`n_iter:` número de combinações a serem selecionadas <i>aleatóriamente</i>; <br>
`cv:` número de `cross-validation`;<br>
`verbose:` Relacionado à quantidade de mensagens que o programa exibe enquanto está em execução;<br>
`random_state:` Para que haja consistência entre as execuções do código, este parâmetro deve ser mantido constante, independente de qual for. Nesse caso, é utilizado o <a href="https://en.wikipedia.org/wiki/The_Hitchhiker%27s_Guide_to_the_Galaxy">42</a>;<br>
`n_jobs:` Quantidade de tarefas sendo executadas em paralelo.`n_jobs=-1` indica que todos os núcleos do processador serão utilizados na realização das tarefas;<br>
`scoring:` meio pelo qual será mensurada a performance de cada configuração gerada pelo `RandomizedSearchCV`;<br>


```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_params, 
                               n_iter = 200, cv = 3, verbose=1, random_state=42, n_jobs = 6, scoring="r2")
```
